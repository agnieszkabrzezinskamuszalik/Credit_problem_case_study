```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel, SelectKBest, RFE, RFECV, VarianceThreshold
titanic_url = ('https://raw.githubusercontent.com/amueller/'
               'scipy-2017-sklearn/091d371/notebooks/datasets/titanic3.csv')
data = pd.read_csv(titanic_url)
data.drop(["body","boat"],axis=1,inplace=True)
data.drop(["name","ticket","home.dest"],axis=1,inplace=True)
data["cabin"] = data.cabin.fillna("X").str[0]
data["sex"] = data.sex.replace({"male":0,"female":1})
data.age.fillna(data.age.mean(),inplace=True)
data.fare.fillna(data.fare.mean(),inplace=True)
data.embarked.fillna(data.embarked.mode()[0],inplace=True)
data = pd.get_dummies(data)
y = data.survived
X = data.drop(["survived"],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=123)
```

```python
model = LogisticRegression('l1', C=0.1)
model.fit(X_train, y_train)
model.coef_
```

```python
accuracy_score(y_test, model.predict(X_test))
```

```python
titanic_url = ('https://raw.githubusercontent.com/amueller/'
               'scipy-2017-sklearn/091d371/notebooks/datasets/titanic3.csv')
data = pd.read_csv(titanic_url)
data.drop(["body","boat"],axis=1,inplace=True)
data.drop(["name","ticket","home.dest"],axis=1,inplace=True)
data["cabin"] = data.cabin.fillna("X").str[0]
data["sex"] = data.sex.replace({"male":0,"female":1})
```

```python
X = data.drop("survived", axis =1)
y = data.survived

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 300)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
num_features = ["pclass", "age", "sibsp", "parch", "fare"]
num_pipeline = make_pipeline(SimpleImputer(strategy = "mean"), 
                             StandardScaler())
cat_feauters = ["sex", "embarked"]
cat_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"),
                            OneHotEncoder(handle_unknown="ignore"))
cabine_pipeline = make_pipeline(SimpleImputer(strategy = "constant", fill_value="X"),
                               OneHotEncoder(handle_unknown="ignore"))
```

```python
from sklearn.compose import ColumnTransformer
```

```python
processor = ColumnTransformer(transformers=[
    ("num", num_pipeline, num_features),
    ("cat", cat_pipeline, cat_feauters),
    ("cabin", cabine_pipeline, ["cabin"])
])
```

```python
classification_process = make_pipeline(processor, LogisticRegression())
```

```python
classification_process.fit(X_train, y_train)
```

```python
classification_process.predict(X_test)
```

```python
accuracy_score(y_test, classification_process.predict(X_test))
```
